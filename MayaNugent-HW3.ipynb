**Exercise**:  An interferometer is a sensitive instrument where a laser beam is split into two pathways and intereference phenomena can inform us about the difference in the length of the two paths.  If we have two paths, one of length $L_1$ and other of length $L_2 + \delta (t)$.  In this example, one path length is constant in time, while the other slowly oscillates, $\delta (t)$.  The electric field of the total electromagnetic wave that reaches the detector is the **imaginary part** of 
$$\tilde{E} (t) = e^{i (k L_1 - \omega t)} + e^{i (k L_2 + k \delta (t) - \omega t)}$$
If we use a red laser with wavelength 670 nm, then $k = 2 \pi / 670~{\rm nm^{-1}}$  and $\omega = 2.8 \times 10^{15}~{\rm s^{-1}}$.  

Lets say that 
$$\delta (t) = A \cos (2 \pi f t)$$
with amplitude $A = 1~{\rm nm}$ and frequency $f = 1~{\rm Hz}$.  The oscillation due to the red light is so incredibly fast compared to the oscillation in $\delta (t)$, we can average the intensity over many periods of the red light without affecting $\delta (t)$, so the time-averaged intensity can be written as
$$I (t) = \tfrac{1}{2} \big\vert \, e^{i k L_1} + e^{i (k L_2 + k \delta (t))} \, \big\vert^2 .$$

We want to plot this intensity as a function of time.  So, let's create a function that has four inputs:
- $L_1$ as measured in meters
- $L_2$ as measured in meters
- $A$ as measured in nanometers
- $f$ as measured in Hertz

Create a function with four inputs, e.g., ``def awesome_HW_function(L1_m, L2_m, A_nm, f_Hz):`` (Please don't steal my awesome function name!), and your function should:

You'll need to note that we're using nm as our length scale, so it'll be useful to transform your lengths into nanometers, e.g., ``L1 = L1_m * 1e9`` and ``L2 = L2_m * 1e9``, to introduce the value of $k$, and to evaluate the expression for $I(t)$.



**Your submission** should create two functions as described above (one for intensity and one to make a graph), then use the function to create **four** graphs.  
- First with ``L1_m = 10``, ``L2_m = 15``, ``A_nm = 1``, and ``f_Hz = 1``.  This should look like a cosine graph with a small amplitude and a frequency of 1 Hz (i.e., 4 complete cycles).
- The other three graphs should have ``L1_m = 10.1``, ``L1_m = 10.2``, and ``L1_m = 10.3``, while keeping the other variables fixed.

```python
def intensity(t,L1_m,L2_m,A_nm,f_Hz):
    return 0.5 * np.abs(np.exp(1j*(2*np.pi/670)*L1_m) + np.exp(1j * ((2*np.pi/670)*L2_m+(2*np.pi/670)*(A_nm*np.cos(2*np.pi*f_Hz*t)))))**2
    
def plot_intensity(L1_m):
    time = np.linspace(0,4,500)
    avg_intensity = np.zeros(len(time))
    
    for i in range(len(time)):
        avg_intensity[i] = intensity(time[i], L1_m*1e9, 15*1e9, 1, 1)
    
    plt.figure()
    plt.plot(time, avg_intensity)
    plt.xlabel("time (s)")
    plt.ylabel("intensity")
    plt.show()

plot_intensity(10)
plot_intensity(10.1)
plot_intensity(10.2)
plot_intensity(10.3)
```
